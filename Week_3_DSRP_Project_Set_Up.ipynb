```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming score_df is already defined as before

# Years to visualize
years = ['2019-2020', '2020-2021', '2021-2022', '2022-2023', '2023-2024', '2024-2025']
np.random.seed(42)  # For reproducibility

# Filter Total MCAT and GPA Total data
total_mcat = score_df[score_df['Metric'] == 'Total MCAT']
gpa_total = score_df[score_df['Metric'] == 'GPA Total']

# Create figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# ------------------ MCAT Histogram ------------------
for year in years:
    mean = total_mcat[total_mcat['Stat'] == 'Mean'][year].values[0]
    sd = total_mcat[total_mcat['Stat'] == 'SD'][year].values[0]
    samples = np.random.normal(mean, sd, 1000)
    axes[0].hist(samples, bins=30, alpha=0.5, label=year)

axes[0].set_title("Total MCAT Scores (Applicants)")
axes[0].set_xlabel("MCAT Score")
axes[0].set_ylabel("Frequency")
axes[0].legend()
axes[0].grid(True)

# ------------------ GPA Histogram ------------------
for year in years:
    mean = gpa_total[gpa_total['Stat'] == 'Mean'][year].values[0]
    sd = gpa_total[gpa_total['Stat'] == 'SD'][year].values[0]
    samples = np.random.normal(mean, sd, 1000)
    axes[1].hist(samples, bins=30, alpha=0.5, label=year)

axes[1].set_title("Total GPA (Applicants)")
axes[1].set_xlabel("GPA")
axes[1].set_ylabel("Frequency")
axes[1].legend()
axes[1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```
